```python
import torch as pt
import matplotlib.pyplot as plt
import numpy as np
```

# MORS - Episode 40


```python
data_ROM = pt.load("mors_final_1/seed0/observations_1.pt")
#data_ROM = pt.load("")
data_CFD = pt.load("mors_final_1/seed0/observations_5.pt")
#data_ROM = pt.load("mors/seed0/observations_39.pt")


#data_CFD = pt.load("/home/tanuj/cylinder_e80_r10_b10_f8_MF_orig/seed0/observations_19.pt")
```

```python
data_ROM[0].keys()
```

```python
plt.plot(data_ROM[0]["rewards"], label = "ROM")
plt.plot(data_CFD[0]["rewards"], label = "MF")
plt.xlabel("time index")
plt.ylabel("rewards")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 4), dpi = 160)
t = np.arange(40, 80.0, 0.1)

plt.plot(t, data_ROM[0]["cx_a"], label = r"Episode $1$")
plt.plot(t, data_CFD[0]["cx_a"], label = r"Episode $5$", linestyle = "--")
plt.title("Drag coefficient", fontsize = 22)
plt.xlabel(r"$\tilde{t}$", fontsize = 22)
plt.ylabel(r"$C_d$", fontsize = 22)
plt.legend(fontsize = 16)
plt.tick_params(axis="both", labelsize=18)
plt.legend(
    loc="upper center",
    bbox_to_anchor=(0.5, -0.18),  # below the plot
    ncol=3,                      # horizontal layout
    fontsize=18,
    frameon=False
)
plt.savefig("drag_comparison.png", bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(6, 4), dpi = 160)
t = np.arange(40, 80.0, 0.1)

plt.plot(t, data_ROM[0]["cy_a"], label = r"Episode $1$", marker = "*")
plt.plot(t, data_CFD[0]["cy_a"], label = r"Episode $5$", linestyle = "--")
plt.title("Lift coefficient", fontsize = 22)
plt.xlabel(r"$\tilde{t}$", fontsize = 22)
plt.ylabel(r"$C_l$", fontsize = 22)
plt.legend(fontsize = 16)
plt.tick_params(axis="both", labelsize=18)
plt.savefig("lift_comparison.png", bbox_inches="tight")
plt.legend(
    loc="upper center",
    bbox_to_anchor=(0.5, -0.18),  # below the plot
    ncol=3,                      # horizontal layout
    fontsize=18,
    frameon=False
)
plt.show()
```

```python
plt.figure(figsize=(6, 4), dpi = 160)
t = np.arange(40, 80.0, 0.1)
print(t.shape)
plt.plot(t, data_ROM[0]["actions"], label = r"Episode $1$")
plt.plot(t, data_CFD[0]["actions"], label = r"Episode $5$", linestyle = "--")
plt.xlabel(r"$\tilde{t}$", fontsize = 22)
plt.ylabel(r"$\omega$", fontsize = 22)
plt.legend(fontsize = 16)
plt.tick_params(axis="both", labelsize=18)
plt.title("Actuation angular velocity", fontsize = 22)

plt.savefig("actions_comparison.png", bbox_inches="tight")
plt.legend(
    loc="upper center",
    bbox_to_anchor=(0.5, -0.18),  # below the plot
    ncol=3,                      # horizontal layout
    fontsize=18,
    frameon=False
)
plt.show()
```
